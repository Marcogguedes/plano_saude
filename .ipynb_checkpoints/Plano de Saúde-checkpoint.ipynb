# 4 Pre-Processamento dos Dados

```python
# Arredondar a varivel idade
train_data['age'] = round(train_data['age'])
```

```python
# Primeiras 5 linhas do dataset depois da realizao do arredondamento
train_data.head()
```

```python
# OHEncoding: Transformando as variveis para o tipo numrico
train_data = pd.get_dummies(train_data, drop_first=True)
```

```python
train_data.head()
```

```python
# colunas do dataset
train_data.columns
```

```python
# reorganizando as colunas para ver melhor
train_data = train_data[['age','sex_male','smoker_yes','bmi','children','region_northwest','region_southeast','region_southwest','charges']]
train_data.head(2)
```

```python
# diviso de recurso independente e dependente
X = train_data.iloc[:, :-1]
y = train_data.iloc[:, -1]
```

```python
# dois melhores registros de recursos independentes
X.head(2)
```

```python
# dois melhores registros de recursos dependentes
y.head(2)
```

```python
# diviso dos dados em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# 4 e 5 Construo e Avaliao de Mquina Preditiva

#### Criao da metodologia de previso do valor do custo do plano de sade que  a prpria mquina preditiva.

```python
# Importando as mtricas de avaliao
from sklearn.metrics import mean_squared_error, r2_score
```

## MP com Regresso Linear

```python
# Linear Regression:
from sklearn.linear_model import LinearRegression
LinearRegression = LinearRegression()
LinearRegression = LinearRegression.fit(X_train, y_train)

# Prediction:
y_pred = LinearRegression.predict(X_test)

# Scores:
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

# MP com Regresso Ridge

```python
# Ridge
from sklearn.linear_model import Ridge
Ridge = Ridge()
Ridge = Ridge.fit(X_train, y_train)

# Prediction:
y_pred = Ridge.predict(X_test)

# Scores:
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

# MP com Regresso Lasso

```python
# Lasso
from sklearn.linear_model import Lasso
Lasso = Lasso()
Lasso = Lasso.fit(X_train, y_train)

# Prediction:
y_pred = Lasso.predict(X_test)

# Scores:
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

# MP com Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
RandomForestRegressor = RandomForestRegressor()
RandomForestRegressor = RandomForestRegressor.fit(X_train, y_train)

# Prediction:
y_pred = RandomForestRegressor.predict(X_test)

# Scores:
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```
